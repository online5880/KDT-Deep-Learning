```python
# YOLOv8 모델 로드
model = YOLO("yolov8s.pt")

# 웹캠 열기 (기본 웹캠은 인덱스 0, 다른 카메라는 인덱스를 조정할 수 있음)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("웹캠을 열 수 없습니다.")
    exit()

# 실시간 스트리밍 시작
while True:
    # 웹캠에서 프레임 읽기
    ret, frame = cap.read()
    if not ret:
        print("프레임을 읽을 수 없습니다.")
        break
    
    # YOLOv8 모델로 프레임 처리 (스트리밍 모드로 실시간 처리)
    results = model(frame, stream=True)
    
    # 결과 시각화 및 출력
    for result in results:
        # 예측된 결과 시각화
        plot = result.plot()

        # OpenCV 창에 결과 출력
        cv2.imshow("YOLOv8 Real-time Detection", plot)
    
    # 'q' 키를 누르면 실시간 스트리밍 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 웹캠 및 OpenCV 창 정리
cap.release()
cv2.destroyAllWindows()
```
