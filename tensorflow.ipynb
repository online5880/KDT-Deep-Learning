````python
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler

print(tensorflow.__version__)
````

````python
mnist = tf.keras.datasets.mnist # 8비트 이미지이다.
````

````python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
````

````python
X_train = X_train/255.0
X_test = X_test/255.0
````

### 딥러닝 모델 만들기

1. 모델 정의(어떤 형태의 모델을 구현할 것인가)

````python
model = tf.keras.models.Sequential(
    [
        tf.keras.layers.Flatten(input_shape=(28, 28)),  # 입력층 
        tf.keras.layers.Dense(128,activation='relu'),   # 입력 레이어, 출력 레이어를 연결해주는 레이어
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(10, activation='softmax') # 출력층
    ]
)
````

2. 모델 컴파일

````python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
````

3. 예측

````python
X_train[0:1]
````

````python
predictions = model(X_train[2:3])
````

````python
predictions.numpy()
````

````python
y_train
````

````python
tf.nn.softmax(predictions).numpy()
````

````python
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
loss_fn(y_train[2:3],predictions).numpy()
````

모델 학습

````python
model.fit(X_train,y_train,epochs=10)
````

````python
model.evaluate(X_test,y_test)
````
